```
data=open("ger_eng_dataset.txt","r")
val=data.read()
pairs=val.split("\n")
len(pairs)
```

```
train=pairs[11000:19000]
test=pairs[10000:10500]
```

```
import re
regex = re.compile('[^a-zA-Z ]')
eng=[]
ger=[]
for pair in train:
  t=pair.split("\t")
  eng.append(regex.sub('',t[0].lower()))
  ger.append(regex.sub('',t[1].lower()))
```

```
ger[10]
```

```
from keras.preprocessing.text import one_hot
from keras.preprocessing.text import text_to_word_sequence
from keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(eng)
```

```
tokenizer.word_index
```

```
Y=tokenizer.texts_to_sequences(eng)
```

```
tokenizer1 = Tokenizer()
tokenizer1.fit_on_texts(ger)
tokenizer1.word_index
```

```
X=tokenizer1.texts_to_sequences(ger)
```

```
X
```

```
a=0
for line in X:
  a=max(a,len(line))
a
```

```
b=0
for line in Y:
  b=max(b,len(line))
b
```

```
from keras.preprocessing.sequence import pad_sequences
X=pad_sequences(X,a,padding='post')
Y=pad_sequences(Y,b,padding='post')
```

```
from numpy import array
import numpy as np
from keras.utils import to_categorical
ylist = list()
for seq in Y:
  encode=to_categorical(seq, num_classes=len(tokenizer.word_index)+1)
  #encode=np.ndarray.tolist(encode)
  #for i in encode:
    #i=np.ndarray.tolist(i)
  
  ylist.append(encode) 

  
y = array(ylist)
y = y.reshape(np.shape(Y)[0],np.shape(Y)[1], len(tokenizer.word_index)+1)
```

```
import numpy as np
np.argmax(y[0][1])
```

```
y[0][1]
```

```
y.shape
```

```
from keras.models import Sequential
from keras.layers import Dense,LSTM,Embedding,Activation,TimeDistributed,Dropout,LeakyReLU,SimpleRNN,Flatten,GRU,RepeatVector,Reshape
```

```
model=Sequential()
#model.add(Dense(a,activation='sigmoid',input_shape=(a,)))

model.add(Embedding(len(tokenizer1.word_index)+1,100,input_length=a,mask_zero=True))
model.add(GRU(150))
model.add(Dropout(0.3))
#model.add(LSTM(40,return_sequences=True))
#model.add(Flatten())
model.add(Dense(100,activation='relu'))
model.add(RepeatVector(b))
model.add(GRU(100,return_sequences=True))


#model.add(Flatten())
model.add(TimeDistributed(Dense(len(tokenizer.word_index)+1, activation='softmax')))

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=["accuracy"])
print(model.summary())
```

```
model.fit(X,y,epochs=35)
```

```
op=model.predict(X)
```
